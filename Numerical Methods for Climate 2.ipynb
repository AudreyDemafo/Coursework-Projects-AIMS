# Exercise 1

Let consider the normalized predator-prey (Lokta Volterra) model given by
\begin{align}
\frac{dB}{dt} & = B-BR \\
\frac{dR}{dt} & = -R + BR
\end{align}

```python
from pylab import*
```

```python
def f1(s,t):
    [b,r] = s
    dbdt = b - b*r
    drdt = -r + b*r
    return np.array([dbdt,drdt])
```

```python
def Euler(x0,t,h):
    s = np.zeros((len(t),len(x0)))
    s[0] = x0
    for i in range(len(t)-1):
        s[i+1] = s[i] + h*f1(s[i],t[i])
    return s
```

```python
def RungeKutta(x0,t,h):
    s = np.zeros((len(t),len(x0)))
    s[0] = x0
    for i in range(len(t)-1):
        k1 = f1(s[i],t[i])
        k2 = f1(s[i]+h*(k1/2),t[i]+(h/2))
        k3 = f1(s[i]+h*(k2/2),t[i]+(h/2))
        k4 = f1(s[i]+h*k3,t[i]+h)
        s[i+1] = s[i] + (h/6)*(k1+2*k2+2*k3+k4)
    return s
```

```python
t = np.linspace(0,20,100)
h = t[1] - t[0]
x0 = [2,1]
sol = Euler(x0,t,h)
```

```python
t = np.linspace(0,20,100)
h = t[1] - t[0]
x0 = [2,1]
sol1 = RungeKutta(x0,t,h)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,4,1)
plt.plot(t,sol[:,0],'-',color = 'red', linewidth = 5)
plt.plot(t,sol[:,1], '--', linewidth = 5, color = 'green')
plt.title('Euler Method')
plt.legend(['Prey(B)','Predator(R)'])
plt.subplot(1,4,2)
plt.plot(t,sol1[:,0],'-',color = 'm', linewidth = 5)
plt.plot(t,sol1[:,1], '--', linewidth = 5, color = 'y')
plt.title('Runge Kutta Method Method')
plt.legend(['Prey(B)','Predator(R)'])
plt.subplot(1,4,3)
plt.plot(sol[:,0],sol[:,1],'-',color = 'k', linewidth = 5)
# plt.plot(sol1[:,0],sol1[:,1],'--', linewidth = 5, color = 'y')
plt.title('Prey(B) vs Predator(R) (Euler)')
# plt.legend(['EulerRunge Kutta'])
plt.subplot(1,4,4)
# plt.plot(sol[:,0],sol[:,1],'-',color = 'm', linewidth = 5)
plt.plot(sol1[:,0],sol1[:,1],'--', linewidth = 5, color = 'y')
plt.title('Prey(B) vs Predator(R) (Runge Kutta)')
# plt.legend(['Euler','Runge Kutta'])
```

# Change of the initial conditions and the time step

```python
t = np.linspace(0,20,1000)
h = t[1] - t[0]
x0 = [2.5,1.1]
sol = Euler(x0,t,h)
```

```python
t = np.linspace(0,20,1000)
h = t[1] - t[0]
x0 = [2.5,1.1]
sol1 = RungeKutta(x0,t,h)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,4,1)
plt.plot(t,sol[:,0],'-',color = 'red', linewidth = 5)
plt.plot(t,sol[:,1], '--', linewidth = 5, color = 'green')
plt.title('Euler Method')
plt.legend(['Prey','Predator'])
plt.subplot(1,4,2)
plt.plot(t,sol1[:,0],'-',color = 'm', linewidth = 5)
plt.plot(t,sol1[:,1], '--', linewidth = 5, color = 'y')
plt.title('Runge Kutta Method Method')
plt.legend(['Prey','Predator'])
plt.subplot(1,4,3)
plt.plot(sol[:,0],sol[:,1],'-',color = 'k', linewidth = 5)
# plt.plot(sol1[:,0],sol1[:,1],'--', linewidth = 5, color = 'y')
plt.title('Prey vs Predator (Euler)')
# plt.legend(['EulerRunge Kutta'])
plt.subplot(1,4,4)
# plt.plot(sol[:,0],sol[:,1],'-',color = 'm', linewidth = 5)
plt.plot(sol1[:,0],sol1[:,1],'--', linewidth = 5, color = 'y')
plt.title('Prey vs Predator (Runge Kutta)')
# plt.legend(['Euler','Runge Kutta'])
```

We have changed the initial conditions of the system as well as the timestep. As we can observe in the two kind of plots, the trajectories  of the variables involved in the system are changing. Even, we can also notice that when we add that small perturbation in the system, the size of both populations are rescaled. For example in the Euler method, the growth rate of prey and predators starts from 0 up to 12. Thenafter changing the initial conditions and the timestep, it start lying between 0 and 2.5.

The change in initial conditions also affects the state of the system in terms of equilibrum.

$\textbf{Difference between Euler method and Runge Kutta method}$

When we have a look at the output that we got from the two methods, we cn notice that the Runge Kutta method gives more approximation for the numerical solution of the above system of ODE's.

For the first plots that we got, we see that with the Runge Kutta method, the interval from 0 up to 2 into 7 with the time step be 0.2. Contraly, for the Euler method, the time step from 0 to 2 is only 1 that is not very accurate.

However, when we change the initial state of the system, thsi accuracy is biased.
